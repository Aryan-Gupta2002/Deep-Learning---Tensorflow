### Creating Random Tensors  
Random Tensors are tensors of some arbitrary size (random) which conatains random numbers  

```python
# Create two random (bt same) tensors
generator1 = tf.random.Generator.from_seed(42)
random1 = generator1.normal(shape=(2,2))
generator2 = tf.random.Generator.from_seed(42) # Same Seed
random2 = generator2.normal(shape=(2,2))
random3 = generator1.uniform(shape=(2,2)) #Uniformaly distributed, not same seed
random1,random2,random3, random1==random2 # same due to same seed
```

### Shuffling the order of elements in our tensor

```python
# Shuffling a tensor is very necessay as to avoid learning too much about the first kind of images (Overfittig)
notShuffled = tf.constant([[[3,7],[8,5],[7,9]],
                          [[4,7],[8,6],[2,1]],
                          [[7,2],[6,5],[4,3]]])
notShuffled.ndim
notShuffled
```

If you want the same order of shuffled tensors , use global + operation level seed

```python
# tf.random.shuffle, shuffles only 1st dimension of a tensor
# Rows shuffles in our case
tf.random.set_seed(42)# Global level seed
tf.random.shuffle(notShuffled,seed=4)# Operational level seed
```

###Turning Numpy arrays to Tensorflow Tensors
The main difference between numpy arrays and tensorflow tensors is that tensors can run on GPUs which is much faster.

```python
import numpy as np
arr = np.arange(1,25,dtype=np.int32)
arr
```

```python
# Converting to tensor
a = tf.constant(arr)
b = tf.constant(arr,shape=(4,3,2)) # Shape should be factor of the nomber of elements in np array
c = tf.constant(arr,shape=(6,4))
a,b,c
```
